#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
sub_z = ds_sub.copy()
datums = []
for i in sub_z.groupby('level5'):
    i[1]['price'] = (i[1]['price'] - i[1]['price'].mean()) / i[1]['price'].std()
    datums.append([i[1]['price'], i[0]])
for i in datums:
    plt.figure(figsize = (10,5))
    plt.hist(i[0], bins = 30, alpha = 0.7)
    plt.title(i[1])
    plt.show()
```

Given the normalization method I used, it becomes abundantly clear that there are a handful of outliers in each district which skew the rest of the data.  Each has at least one house with a standard deviation between 4 and 6, sitting on the high side.  This is far enough away from the rest of the houses in the population, that it pushes the bulk, which should be centered around 0 to closer to -1.  The one exception to this is only an exception in the fact that it has a second peak sitting just below 1 deviation, indicating a middle class and a slightly more padded distribution (Getafe).

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in the question 14) (★★★★)

Execute the following cell and then start coding in the next one. You must implement a simple code that transform the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
import folium
m = folium.Map(location = [40.317715, -3.784103393554688], zoom_start= 10)
m
```

```python
## HERE: plot the coordinates of the estates
dic = {}
for i in ds_sub.groupby('level5')[['latitude', 'longitude']]:
    dic[i[0]] = {'lat':[], 'long':[]}
    for j in i[1]['latitude']:
        dic[i[0]]['lat'].append(j)
    for j in i[1]['longitude']:
        dic[i[0]]['long'].append(j)
## PUT HERE YOUR CODE:
locales = folium.map.FeatureGroup()
colind = 0
colors = ['red', 'blue', 'green', 'black']
for k in dic.keys():
    for i in range(len(list(dic[k].values())[0])):
        dic[k]['lat'][i] = dic[k]['lat'][i].replace(',', '.')
        dic[k]['long'][i] = dic[k]['long'][i].replace(',', '.')
        locales.add_child(folium.CircleMarker([dic[k]['lat'][i], dic[k]['long'][i]], radius = 5, stroke = False, fill = True, fill_opacity = 0.5, color = colors[colind], opacity = 0.2))
    colind += 1
m.add_child(locales)

##Formatting isn't working for some reason.  Apologies for ugly circle markers.
```
